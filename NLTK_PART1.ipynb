### Removing stop Words from the text

```python
text # this is my text on which I have to remove stopwords
```

```python
stop_words = set(stopwords.words('English'))
```

```python
stop_words
```

```python
text_to_word_tokenize = word_tokenize(text)
```

```python
new_words = []
```

```python
for t in text_to_word_tokenize:
    if t not in stop_words:
        new_words.append(t.lower())
```
